```python
import datetime

import pandas as pd
```

```python
csvs = ['viajes_2012.csv', 'viajes_2015.csv', 'viajes_2016.csv', 'viajes_2017.csv']

for cnt, csv in enumerate(csvs):
    if cnt == 0:
        df = pd.read_csv(csv)
        fecha = csv.split('_')[1].split('.csv')[0]
        df['Año'] = [fecha for _ in range(df.shape[0])]
    else:
        df_temp = pd.read_csv(csv)
        fecha = csv.split('_')[1].split('.csv')[0]
        df_temp['Año'] = [fecha for _ in range(df_temp.shape[0])]
        
        df = pd.concat([df, df_temp], ignore_index=True)
        
df
```

```python
def to_datetime(ele):
    if str(ele) != 'nan':
        return datetime.datetime.strptime(str(ele), '%d-%m-%Y')
    else:
        return ele
    
def get_days(ele):
    try:
        return ele.days
    except:
        return ele

df['Fecha_llegada'] = df['Fecha_llegada'].apply(to_datetime)
df['Fecha_salida'] = df['Fecha_salida'].apply(to_datetime)

df['Duracion_viaje'] = df['Fecha_llegada'] - df['Fecha_salida']
df['Duracion_viaje'] = df['Duracion_viaje'].apply(get_days)

df.head()
```

```python
df.to_csv('viajes.csv', index=False)
```
